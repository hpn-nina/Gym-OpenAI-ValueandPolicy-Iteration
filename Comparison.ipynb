```python
import scipy
import gym
import numpy as np
from IPython import display
import time
import datetime 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd 
import random
from scipy.stats import norm
import statistics
```

<h1>This is the drawing part of FrozenLake-v0</h1>

```python
dfp = pd.read_csv('./Success/Success-Policy-FrozenLake-v0.txt',header=0, index_col=None, names=['Success'])
dfv = pd.read_csv('./Success/Success-Value-FrozenLake-v0.txt', header=0, index_col=None, names=['Success'])
#Things you need to do here is to read the number of success time of both algorithm and draw a normal distribution
#Prepare stats for normal distribution plot
successPolicy = dfp["Success"].to_numpy()
muP, stdP = norm.fit(successPolicy)

plt.hist(successPolicy,color='purple',bins=50, density=True)
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = norm.pdf(x,muP,stdP)
plt.ylabel('Density')
plt.xlabel('Success')
plt.title('Success Distribution FrozenLake - Policy Iteration - mu=' + str(muP) + ' std=' + str(stdP))
plt.plot(x, p, 'k', linewidth=2)
plt.tight_layout()
plt.savefig('./Success/FL-Policy.png', dpi=300, bbox_inches='tight')
plt.show()
plt.clf()


successValue = dfv["Success"].to_numpy()
muV, stdV = norm.fit(successValue)
plt.hist(successValue, color='blue', bins=50, density=True)
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = norm.pdf(x,muV,stdV)
plt.title('Success Distribution FrozenLake - Value Iteration - mu=' + str(muV) + ' std=' + str(stdV))
plt.ylabel('Density')
plt.xlabel('Success')
plt.plot(x,p, 'k', linewidth=2)
plt.tight_layout()
plt.savefig('./Success/FL-Value.png', dpi=300, bbox_inches='tight')
plt.show()
plt.clf()
```
